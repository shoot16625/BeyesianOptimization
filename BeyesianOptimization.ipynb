```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import GPy
from scipy.stats import norm
%matplotlib inline
```

```python
# 実測データの生成
x1 = np.random.rand(100)
x2 = np.random.rand(100)
y = np.sin(x1*2*np.pi)*np.sin(x2*2*np.pi)

plt.scatter(x1, x2, c=y, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# ガウス過程回帰モデル
def model_generate(x, y):
    kernel = GPy.kern.Matern52(2, ARD=True)
    model = GPy.models.GPRegression(x, y, kernel)
    model.optimize()
    return model

x = np.c_[x1, x2]

model = model_generate(x, y.reshape(-1, 1))
model.plot()
```

```python
# 獲得関数
def EI(mean, var, point):
    # EI(x)=(μ(x)−f−ξ)Φ(Z)+σ(x)ϕ(Z)　ξは0.01くらいが良い
    Z = (mean - point) / var
    z = np.array([(mean[i] - point - 0.01) * norm.cdf(Z[i]) +
                  var[i] * norm.pdf(Z[i]) for i in range(len(Z))])
    return z


def PI(mean, var, point):
    Z = (mean - point - 0.01) / (var)
    z = np.array([norm.cdf(Z[i]) for i in range(len(Z))])
    return z


def UCB(mean, var):
    return mean + 1.0 * var

# 各格子点におけるスコア計算
def get_score(model, y_sample):
    pred_mean, pred_var = model.predict(x_search)  # 各格子点における期待値と分散を算出
    z_max = max(y_sample)
    theta = 10000  # 分散が小さすぎるとオーバーフローするので，適当にthetaを設定
    z = EI(pred_mean, pred_var * theta, z_max)
    return pred_mean, z

def display(top_n):
    index = z.argsort()[-1::-1]
    Input = x_search[index[:top_n]]
    Output = pred_mean[index[:top_n]]
    x_target_display = pd.DataFrame(Input)
    x_target_display['y_predict'] = Output
    return x_target_display
```

```python
# 格子点生成
a = np.arange(0, 1, 0.01)
X, Y = np.meshgrid(a, a)
X = X.reshape(-1)
Y = Y.reshape(-1)
x_search = np.c_[X, Y]

pred_mean, z = get_score(model, y)
pred_mean = pred_mean.reshape(-1)
z = z.reshape(-1)
```

```python
# 探索価値の高い上位top_nを表示（期待値：大＆分散：大）
top_n = 10
display(top_n)
```
